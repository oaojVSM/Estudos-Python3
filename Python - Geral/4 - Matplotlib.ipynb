Bom, agora vamos falar de fato de como plotar algum gráfico. O matplotlib trabalha muito bem com Pandas e Numpy. Então, agora importarei Numpy para criar um array que utilizaremos como fonte de dados para os nossos gráficos. Veja que criarei basicamente uma função de segundo grau:

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

Primeiramente, eu vou direto ao ponto e plotar o gráfico da forma mais funcional e simples. Mas já noto que muitas pessoas não tem preferência em realizar a plotagem dessa maneira pois isso prejudica maiores controles que você tem sobre a figura. Depois mostrarei a forma mais completa de plotar um gráfico. <br><br> 

A forma mais funcional é por meio da função **plt.plot**

```python
help(plt.plot)
```

```python
#Veja como pode parecer fácil:
plt.plot(x,y)
```

```python
#agora vou realizar o mesmo plot porém adicionando algumas informações
plt.plot(x,y)
plt.title('Primeiro Gráfico')
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do eixo Y')
```

Podemos plotar 2 gráficos de uma vez por meio dos **subplots**:

```python
help(plt.subplot)
```

```python
plt.subplot(1,2,1) # veja a argumentação que coloquei acima no docstring da função
plt.plot(x,y,'r') #o "r" é a cor da linha, falaremos mais disso posteriormente
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

Agora, vou mostrar como podemos realizar a plotagem de gráficos por meio de uma maneira mais orientada ao objeto, mais completa e que nos permitirá uma maior manipulação dos elementos nos gráficos:

```python
graf = plt.figure() # aqui eu criei uma figura em branco, imagine que seja a base de um gráfico, mas que nao contém nada
eixos = graf.add_axes([0.1,0.1,0.8,0.8]) # aqui eu criei os eixos na figura previamente em branco, veja o help dessa função

eixos.plot(x,y) # coloquei os dados nos eixos
eixos.set_title('Gráfico Orientado ao Objeto')
eixos.set_xlabel('Eixo X') # note que eu estou começando os métodos com set_
eixos.set_ylabel('Eixo Y')
```

```python
help(plt.figure.add_axes)
```

Vamos falar um pouco do método **add_axes**, que foi algo que eu tive um pouco de dificuldade de entender no começo.<br><br>

Quando criei o objeto *graf*, na verdade, ele não é um gráfico em branco, mas sim uma figura em branco, pense como se fosse uma folha de papel sem nada escrito na qual posteriormente vamos incluir nosso gráfico. Beleza, após isso, a primeira coisa que fazemos é determinar onde colocar os eixos do gráfico nessa folha de papel (são determinados pelos dois primeiros elementos da lista que coloquei como argumento), e também determinamos o tamanho desses eixos (são determinados pelos dois últimos elementos da lista que coloquei como argumento). <br><br> 

Veja essa ilustração retirada do site _[TowardsDataScience](https://towardsdatascience.com/)_ que pode ser elucidativa:

O que podemos fazer também é colocar mais de uma gráfico dentro da mesma figura. Pense literalmente na figura como uma folha de papel e podemos fazer o gráfico da forma que quisermos!

```python
papel = plt.figure()

#colocando o primeiro gráfico:
graf1 = papel.add_axes([0.05,0.05,0.9,0.8])
graf1.plot(x, y,'r')
graf1.set_title('Gráfico Maior')
graf1.set_xlabel('Eixo X do Gráfico Maior')
graf1.set_ylabel('Eixo Y do Gráfico Maior')

#colocando o segundo gráfico:
graf2 = papel.add_axes([0.1,0.4,0.4,0.3])
graf2.plot(y, x, 'b')
graf2.set_title('Gráfico Menor')
plt.show()
```

Uma outra forma ainda, que se assemelha à anterior que podemos utilizar para pltar gráficos é por meio da função **subplots**, que pula algumas etapas e já cria a figura com os gráficos de forma automática. Veja:

```python
help(plt.subplots)
```

Veja que está função retorna uma figura e os eixos ao mesmo tempo, portanto utilizamos tuple unpacking. Vou fazer um exemplo e talvez fique mais simples:

```python
fig, graf = plt.subplots()

graf.plot(x,y)
```

Com essa função, podemos plotar mais de um gráfico de maneira organizada:

```python
fig, graf = plt.subplots(nrows=1,ncols=2)
```

```python
type(graf)
```

Veja que a variável *graf* nada mais é do que um Numpy Array! Isso significa que os gráficos estão organizados de forma como se fossem uma lista, e eu para eu plotar informações em cada gráfico, eu posso indexá-los como se fosse uma lista, veja:

```python
fig, graf = plt.subplots(nrows=1,ncols=2)
graf[0].plot(x,y,'r')
graf[0].set_title('Gráfico da Esquerda')

graf[1].plot(y,x,'b')
graf[1].set_title('Gráfico da Direita')
# poderia usar o set_ylabel e xlabel normalmente em cada um dos gráficos
plt.show()
```

Uma vez que podemos indexar os gráficos como se fossem elementos de uma lista, podemos iterar sobre eles!! Veja:

```python
fig, grafs = plt.subplots(nrows=3,ncols=1,figsize=(12,6),dpi=400) #posso alterar figsize para melhor acomodar o meu plot

for graf in grafs:
    graf.plot(x,y,'purple')
    graf.set_title('Gráfico')
plt.tight_layout() #isso impede que as figuras fiquem "overlapping", experimente rodar o código sem esse comando
```

#### Legenda

Vamos ver como podemos colocar legendas nos gráficos, também:

```python
fig, graf = plt.subplots()

graf.plot(x,y,label='X ao Quadrado') #preciso colocar o argumento "label"
graf.legend() #depois preciso inserir esse comando, como argumento, posso especificar a posição da legenda, veja o help
```

#### Formatação

Podemos alterar de diversas maneiras o nosso plot, colocando diferentes tipos de linhas, adicionando marcadores, mudando as cores. Isso tudo pode ser feito por meio de adicionar argumentos na função **plot**:

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25) # na cor eu posso colocar o RGB Hex Code para maiores customizações
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
