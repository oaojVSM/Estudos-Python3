# Teoria de Portfolio de Markowitz

Neste notebook, vou introduzir o conceito da Teoria Moderna de Portfolios de Markowitz, uma das principais teorias de finanças modernas!

A ideia por trás do entendimento da teoria de Markowitz é entender a necessidade de diversificação, então, para prosseguir, sugiro fortemente que você primeiro pesquise um pouco e entenda toda a questão da diversificação do um portfolio de ativos.

O que Henry Markowitz fez, em 1952, foi provar a existência de uma fronteira eficiente de portfolios. E o que isso significa?

Bom, basicamente significa que ele criou uma forma de calcular a distribuição ótima dos ativos dentro de uma carteira de investimentos para que você possa ter a diversificação mais eficiente possível considerando Risco x Retorno! Ou seja, para cada nível de risco, seria possível calcular o portfolio com o maior nível de retorno possível.

Tudo isso vai ficar mais claro quando de fato começarmos a calcular tudo isso em breve! Portanto, se mais delongas, vamos começar a importar as libs necessárias e realizar os cálculos e irei comentando ao longo do percurso:

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb #lib que utilizamos pra puxar os dados do Yahoo Finance
from pandas.util.testing import assert_frame_equal
import matplotlib.pyplot as plt
import seaborn as sns
```

Primeiramente, vamos importar alguns ativos para trabalhar:

```python
ativos = ['B3SA3.SA', "VALE3.SA"] #lista com os ativos que vamos importar

pf = pd.DataFrame() #DF vazio que utilizaremos para inserir os preços dos ativos

for ativo in ativos:
    pf[ativo] = wb.DataReader(name=ativo, 
                              data_source='yahoo', 
                              start='2020-1-1', 
                              end='2021-1-1'
                             )['Adj Close'] #crio colunas no DF pf com o fechamento dos ativos
```

```python
pf
```

```python
pf.describe()
```

```python
(pf / pf.iloc[0] * 100).plot(figsize=(10, 5)) #plotando os preços normalizados em base 100
```

```python
log_returns = np.log(pf / pf.shift(1))
#calculando os log retornos em uma variável log_returns
```

```python
log_returns.mean() * 250
#dando uma olhada no retorno médio esperado
```

```python
log_returns.cov() * 250
#dando uma olhada na matriz de covariância dos ativos
```

```python
log_returns.corr()
# mesma coisa para correlação
```

Beleza, agora que já temos nossos dados e demos uma olhada neles, vamos começar a realizar os cálculos que chegarão eventualmente no resultado de obter a fronteira eficiente de Markowitz.

A primeira coisa que tenho que fazer, é calcular possíveis distribuições entre os ativos da B3 e da Vale em uma carteira de ações. E, concordemos que temos **infinitas** distribuições. E, o que farei será o seguinte:

```python
qtd_ativos = len(ativos)
# Armazeno a quantidade de ativos em uma variável
```

```python
qtd_ativos
```

```python
pesos = np.random.random(qtd_ativos) # Calculo dois pesos de forma aleatória
pesos /= np.sum(pesos) # Faço essa conta para que a soma dos pesos seja igual a 1!
pesos # Aqui estou apenas retornando os pesos calculados
```

Acima temos o cálculo de dois pesos aleatórios para cada ativo dentro do meu portfolio! Sendo que essa soma sempre dará 1!

Eventualmente, para que consigamos de fato calcular a fronteira eficiente, eu colocarei esse cálculo de pesos em loop, rodando diversas vezes, para que consigamos calcular diversas distribuições e de fato consigamos "preencher uma fronteira"!

Agora vamos criar 4 listas vazias, contendo o retorno esperado de um portfolio, a volatilidade esperada, o peso do ativo 1 (B3SA3) e do ativo 2 (VALE3):

```python
pf_ret = []
pf_vol = []
peso_B3SA3 = []
peso_VALE3 = []
```

Agora vou criar um loop com 1500 iterações, que irá gerar 1500 portfolios com pesos aleatórios, junto com sua volatilidade e retorno esperados:

```python
for x in range(1500): #loop
    
    pesos = np.random.random(qtd_ativos)
    pesos /= np.sum(pesos)
    pf_ret.append(np.sum(pesos*log_returns.mean()*252)) # aqui calcula o retorno esperado de cada portfolio
    pf_vol.append(np.sqrt(np.dot(pesos.T, np.dot(log_returns.cov()*250, pesos)))) # aqui calcula a volatilidade de cada portfolio
    peso_B3SA3.append(pesos[0]) # aqui eu coloco o peso de B3
    peso_VALE3.append(pesos[1]) # aqui eu coloco o peso de Vale    
```

```python
pf_ret = np.array(pf_ret)
pf_vol = np.array(pf_vol)
peso_B3SA3 = np.array(peso_B3SA3)
peso_VALE3 = np.array(peso_VALE3)

# Aqui estou apenas transformando as listas em vetores para colocar tudo em um novo DF
```

```python
portfolios = pd.DataFrame({
    'Retornos' : pf_ret,
    'Volatilidades' : pf_vol,
    '% B3' : peso_B3SA3,
    '% Vale' : peso_VALE3
}) #criei um DF com as infos que calculei
portfolios
```

Agora, finalmente, vamos plotar a nossa fronteira eficiente!

```python
sns.scatterplot(y=portfolios.Retornos, x=portfolios.Volatilidades)
```

Acima temos a nossa fronteira eficiente!

Se quisermos, por exemplo, pegar o nosso portfolio de mínima variância, que seria "o ponto máximo esquerdo da nossa fronteira", por exemplo, podemos fazer o seguinte:

```python
portfolios[(portfolios.Volatilidades == portfolios.Volatilidades.min())]
# Aqui estou filtrando a linha do DF onde a volatilidade é a menor calculada nas 1500 possibilidades!
```

Legal, conseguimos calcular a fronteira eficiente de Markowitz e verificar o portfolio de mínima variância para uma carteira de 2 ativos! Porém, e se tivermos 3 ativos, ou 4, ou 15?

A lógica é absolutamente a mesma! Para exemplificar, vou mostrar rapidamente um exemplo para 4 ativos:

```python
ativos = ['B3SA3.SA', 'VALE3.SA', 'OIBR4.SA', 'MELI34.SA'] #coloquei um BDR só para dar uma diversificada

pf = pd.DataFrame() #DF vazio que utilizaremos para inserir os preços dos ativos

for ativo in ativos:
    pf[ativo] = wb.DataReader(name=ativo, 
                              data_source='yahoo', 
                              start='2020-1-1', 
                              end='2021-1-1'
                             )['Adj Close'] #crio colunas no DF pf com o fechamento dos ativos
```

```python
pf
```

```python
pf.describe()
```

```python
(pf / pf.iloc[0] * 100).plot(figsize=(10, 5)) #plotando os preços normalizados em base 100
```

```python
log_returns = np.log(pf / pf.shift(1))
#calculando os log retornos em uma variável log_returns
```

```python
log_returns.corr()
```

```python
qtd_ativos = len(ativos)
# Armazeno a quantidade de ativos em uma variável
```

```python
qtd_ativos
```

```python
pf_ret = []
pf_vol = []
peso_B3SA3 = []
peso_VALE3 = []
peso_OIBR4 = []
peso_MELI34 = []
```

```python
for x in range(3000): #loop
    
    pesos = np.random.random(qtd_ativos)
    pesos /= np.sum(pesos)
    pf_ret.append(np.sum(pesos*log_returns.mean()*252)) # aqui calcula o retorno esperado de cada portfolio
    pf_vol.append(np.sqrt(np.dot(pesos.T, np.dot(log_returns.cov()*250, pesos)))) # aqui calcula a volatilidade de cada portfolio
    peso_B3SA3.append(pesos[0]) # aqui eu coloco o peso de B3
    peso_VALE3.append(pesos[1]) # aqui eu coloco o peso de Vale
    peso_OIBR4.append(pesos[2]) # peso de OIBR4
    peso_MELI34.append(pesos[3]) # peso de MELI34
```

```python
pf_ret = np.array(pf_ret)
pf_vol = np.array(pf_vol)
peso_B3SA3 = np.array(peso_B3SA3)
peso_VALE3 = np.array(peso_VALE3)
peso_OIBR4 = np.array(peso_OIBR4)
peso_MELI34 = np.array(peso_MELI34)
```

```python
portfolios = pd.DataFrame({
    'Retornos' : pf_ret,
    'Volatilidades' : pf_vol,
    '% B3' : peso_B3SA3,
    '% Vale' : peso_VALE3,
    '% Oi' : peso_OIBR4,
    '% Meli' : peso_MELI34
}) #criei um DF com as infos que calculei
portfolios
```

```python
sns.scatterplot(y=portfolios.Retornos, x=portfolios.Volatilidades)
```

```python
portfolios[(portfolios.Volatilidades == portfolios.Volatilidades.min())]
```

## Trazendo o conceito do índice Sharpe:

Uma outra métrica que conversa muito bem com o conceito da fronteira eficiente de Markowitz é o índice Sharpe, criado por William Sharpe, nos permitindo avaliar a relação entre o retorno e o risco de um investimento, e é dado pela seguinte fórmula:

**Índice Sharpe:**
### $$
Sharpe = \frac{\overline{r_{i}} - r_f}{\sigma_{i}}
$$

Leia-se: o índice Sharpe é igual à razão da diferença entre o retorno esperado de um ativo (ou portfólio) e a taxa livre de risco, e o risco desse ativo (ou portfólio), que é representado pela volatilidade (desvio padrão).

Vamos pensar no índice Sharpe de forma mais intuitiva: é razoável assumir que investidores buscam maximizar seus retornos pelo menor risco possível, e o Sharpe nos traz uma métrica que torna tangível o cálculo de um "retorno ajustado pelo risco" para que possamos verificar de forma prática, aliando esse indicador à fronteira de Markowitz, qual seria o melhor portfólio dentro da fronteira! Sendo assim, buscamos o portfólio que tenha o maior índice de Sharpe!

O portfólio com maior índice Sharpe seria o portfólio que permite o maior nível de retorno pelo menor nível de risco!

Para aliar o índice Sharpe à nossa análise da fronteira eficiente de Markowitz, vamos primeiramente adicionar uma coluna que irá conter o índice:

```python
portfolios['sharpe'] = (portfolios.Retornos - 0.035) / portfolios.Volatilidades
```

```python
portfolios
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y=portfolios.Retornos, x=portfolios.Volatilidades, c=portfolios.sharpe)
plt.colorbar(label='Sharpe')
plt.xlabel('Volatilidade')
plt.ylabel('Retorno')
plt.show()
```

Acima eu plotei o mesmo gráfico que estávamos vendo anteriormente, porém ele contém uma informação adicional que é índice Sharpe, se apresentando por meio de um *colormap*. E, agora conseguimos ver qual seria o portfolio que possui o maior Sharpe dentre os portfolios gerados:

```python
portfolios[(portfolios.sharpe == portfolios.sharpe.max())] # Portfolio com maior Sharpe da fronteira de Markowitz!
```

```python
# Vamos armazenar o retorno e a vol desse portfolio para destacá-lo no gráfico:

ret_sharpe = portfolios[(portfolios.sharpe == portfolios.sharpe.max())]['Retornos']
vol_sharpe = portfolios[(portfolios.sharpe == portfolios.sharpe.max())]['Volatilidades']

print(
'O portfolio que possui o maior índice sharpe possui um retorno de {ret} e volatilidade de {vol}'.format(
    ret = "%.2f%%" % (100 * ret_sharpe),
    vol = "%.2f%%" % (100 * vol_sharpe)
))
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y=portfolios.Retornos, x=portfolios.Volatilidades, c=portfolios.sharpe)
plt.colorbar(label='Sharpe')
plt.xlabel('Volatilidade')
plt.ylabel('Retorno')
plt.scatter(vol_sharpe, ret_sharpe, c='red', s=70) # Colocando o marker do PF com maior Sharpe em destaque
plt.show()
```
