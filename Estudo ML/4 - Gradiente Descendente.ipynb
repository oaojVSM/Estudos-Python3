# Gradiente Descendente

Trabalhando com problemas de modelagem, muitas vezes teremos que resolver problemas do tipo: 

*- Qual é a abordagem ótima para este problema?*

E muitas vezes esse problema é resolvido por meio de minimizar (ou maximizar) alguma função!

É para isso que existe a técnica do **gradiente descendente** (existe também a técnica do gradiente ascendente, para maximizar alguma função), que se trata de uma técnica, ou melhor dizendo, um algoritmo, que irá minimizar uma determinada função!

Se trata de um algoritmo iterativo, que, em termos simples, irá caminhar por alguma função, até encontrar um ponto de mínimo!

## Explicando o gradiente e a derivada...

Gradiente é simplesmente um outro jeito de chamar a derivada parcial. **Se trata do vetor das derivadas parciais de uma determinada função!**

Caso você nunca tenha tido contato com cálculo na sua vida, isso pode ter parecido meio confuso então vou **tentar** explicar do que se trata. *Note a ênfase no **tentar**, porque este assunto é relativamente avançado e, pelo menos pra mim, foi um pouco difícil de entender no primeiro contato*.

A derivada de uma função pode ser interpretada de duas formas, que, na minha opinião, são as mais intuitivas:

- É a inclinação da reta (ou plano, etc) tangente à função em um determinado ponto.

- É a variação instantânea de uma função em um determinado ponto.

Vou tentar explicar o que eu quis dizer com isso, e porque essas explicações são as mais intuitivas, e também, porque elas são meio que a mesma coisa.

*Obs.: eu não vou tentar ensinar como calcular uma derivada, apenas a intuição por trás, de forma simples. Caso tenha curiosidade ou queira aprender mais a respeito, o que eu recomendo, sinta-se livre para pesquisar vídeos no YouTube sobre o assunto!*

Um exemplo talvez mais prático e clássico na explicação da derivada é a seguinte: A derivada da velocidade em função do tempo é a aceleração! Ou seja, é a unidade de medida pela qual a velocidade muda em função do tempo! Justamente porque a derivada vai nos retornar uma função que é a variação da função da velocidade, ou seja, nos retornar a aceleração.

Podemos também ter uma abordagem um pouco mais algébrica para entender uma derivada:

Vamos pensar em uma das funções mais simples possíveis, a de uma reta:

```python
# Criando e plotando uma função de reta...

# Primeiro importo as libs que eventualmente devo precisar:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crio a função e ploto uma função que se assemelha a isso: f(x) = x

x = np.linspace(0,150,200)
plt.plot(x*5, x)
```

Acima eu plotei uma equação de reta das mais simples possíveis:

$$f(x) = \frac{1}{5}x$$

Se trata de uma função de primeiro grau, que, por consequência possui um número absoluto como derivada, ou seja, ela possui a mesma derivada independentemente do ponto no qual estamos olhando a função, o que não é o caso para função de grau maior do que 1.

Caso não tenha ficado claro, você pode tentar entender essa função da seguinte forma:

E bom, antes de prosseguir, eu faço o seguinte questionamento... Caso você tente desenhar uma reta tangente à que está no gráfico acima, qual seria sua inclinação? E, uma outra pergunta, se você tivesse que chutar "a variação instantânea" dessa função, qual seria?

Olha... Eu tentei tornar o mais intuitivo possível (apesar de ter certeza que eu não consegui) de que a resposta para as duas perguntas acima é 1/5!

Isso se dá porque a reta tangente à uma função de reta, é a própria reta! E, no caso, a inclinação da reta acima é 1/5 (ou 0,2)! E, caso tentemos calcular a variação, ou seja, o quanto *y* varia em função de *x* (podemos pensar literalmente no $\Delta$X/$\Delta$Y das aulas de física da escola!), teremos que a resposta é a mesma!

*Obs.: esse $\Delta$X/$\Delta$Y nada mais é do que de fato a derivada de uma função!*

Se você ainda preferir o exemplo do carro, pense da seguinte forma:

```python
# Crio a função e ploto uma função que se assemelha a isso: f(x) = x

x = np.linspace(0,150,200)
plt.plot(x*5, x)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (km/h)')
```

Responda a seguinte pergunta: o quanto a velocidade varia em função do tempo (em outras palavras, queremos a aceleração), e podemos ver que a cada segundo, a velocidade está variando em 0,2 km/h, ou seja, temos uma aceleração de 0,2 km/h/s. Que é o mesmo valor de nossa derivada!

## Funções de maior grau

A derivada começa a ficar muito mais interessante quando "subimos" o grau das nossas funções, porque conseguimos utilizá-la para econtrar pontos de mínimo e máximo de funções!

```python
# Tipo isso:

plt.plot(x, x**2)
```

Tente entender que agora, a nossa derivada não será mais um número absoluto, porque a inclinação da reta tangente, e a variação instantânea da função irá variar conforme "andarmos" pela parabola! Ou seja, nesses casos, a derivada irá nos retornar uma outra função!

A derivada, neste caso, será igual à: $f'(x) = 2x$ <br><br>

*Obs.: o " ' " na depois do " f " indica que estamos falando da derivada, no caso, da primeira derivada da função*.

Ou seja, conforme o valor de $x$, a derivada terá um valor diferente, porque estaremos falando de um ponto na função no qual a inclinação da reta tangente será diferente, e consequentemente, a variação instantânea também.

## Derivando e igualando a zero...

Eu disse anteriormente que a derivada se torna interessante em funções de graus maiores porque ela se mostra como um meio de encontrar o máximo e o mínimo de alguma função, ou seja, por meio dela, conseguimos maximizar ou minimizar algo!

E isso é feito da seguinte forma: derivando e igualando a zero! Sim, é tão simples assim! E isso funciona pelo seguinte motivo. Pense que você está no ponto inferior da parábola, ali onde $x = y = 0$

Neste exato ponto, a reta tangente à função não possui inclinação positiva nem negativa, ela é uma reta horizontal, ou seja, de inclinação 0, e como corolário, a variação instantânea da função é nula!

Ou seja, se igualarmos a derivada da função à 0, encontraremos que o ponto $x$ mínimo é 0! E, ao aplicar $x = 0$ na função,  você consegue descobrir que o $y$, ou $f(x)$ também é igual à 0!

*** **Disclaimer** ***: talvez eu tenha simplificado demais quando eu disse que basta derivar e igualar à zero para minimizar (ou maximizar) uma função. A verdade é que sim, isso vai funcionar para muitas funções, mas também não vai funcionar para muitas outras. Porque, me adiantando em um tópico um pouco mais avançado, existem funções que possuem os chamados pontos de máximo e mínimo locais e globais. Mas, por ora, e para efeitos da explicação do gradiente descendente, irei abordar apenas funções que possuem um ponto de mímino.

## Utilizando o gradiente descendente...

Vamos agora, finalmente, abordar a técnica do gradiente descendente, para minimizar uma função de perda.

Para isso, vou me utilizar de uma abordagem, simplificada, mas muito semelhante, com partes adaptadas do [excelente vídeo do ritvikmath](https://www.youtube.com/watch?v=xOB10eTjoQ8), e outras partes que considero mais adequadas, na minha humilde opinião.

Vamos supor que estamos tentando criar um modelo que vai replicar o comportamento da seguinte amostra:

```python
numero_de_pontos = 100

x = np.sort(5*(np.random.random(numero_de_pontos) - 0.5))
y = (x + x**2) + np.random.normal(0, 0.1, numero_de_pontos)

plt.figure(figsize=(8,8))
plt.scatter(x, y)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.title('Amostra', fontsize=20)
```

E esse modelo vai seguir a seguinte função:

$$\hat{y} = w (x + x^2)$$

E, a nossa função de perda nada mais é do que do que a soma dos erros quadráticos:

$$ L(y,\hat{y}) = \sum_{i=1}^{n}(\hat{y}_i - y_i)^2$$

Podemos facilmente substituir o $\hat{y}$ por $w$ e $x$:

$$ L(y,x,w) = \sum_{i=1}^{n}(wx_i + wx_i^2 - y_i)^2$$

E, se derivarmos essa função de perda em função de $w$, teremos:

$$ \frac{dL}{dw} = \sum_{i=1}^{n}2(x_i + x_i^2)(wx_i + wx_i^2 - y_i)$$

```python
# Podemos ver que o nosso modelo realmente parece seguir o comportamento da amostra

plt.plot(x, x+x**2)
```

Caso não seja claro, o que queremos descobrir é o $w$ que irá minimizar a nossa função de perda!

Primeiro, o que vou fazer é transformar as funções acima em funções do Python:

```python
def perda(w, x, y):
    return np.sum((w*x + w*x**2 - y)**2)
```

```python
def gradiente(w, x, y):
    return np.sum(2*(x + x**2)*(w*x + w*x**2 - y)) 
```

Agora podemos dar uma olhada na cara da função de perda:

```python
wvals = np.arange(0,1.5,0.01)
perdas = [perda(w,x,y) for w in wvals]

plt.figure(figsize=(12,4))
plt.plot(wvals, perdas)
plt.xlabel('w', fontsize=15)
plt.ylabel('Perda', fontsize=15)
plt.title('Função de Perda', fontsize=20)
```

Visualmente, conseguimos perceber que a função de perda possui um mínimo quando o $w$ está em torno de 1... Mas, **em matemática não se trabalha com aproximações visuais** *(até porque eventualmente podemos trabalhar com funções de maiores dimensões, que não poderiam ser visualizadas em gráficos)*, e iremos calcular esse mínimo por meio do gradiente descendente...

Podemos também dar uma olhada no gradiente, que nada mais é do que a derivada parcial dessa função em torno de $w$:

```python
grad = [gradiente(w, x, y) for w in wvals]

plt.figure(figsize=(12,4))
plt.plot(wvals, grad)
plt.axhline(0, color='black')
plt.xlabel('w', fontsize=15)
plt.ylabel('Valor do Gradiente', fontsize=15)
plt.title('Gradiente', fontsize=20)
```

Note que, olhando o graidente, conseguimos confirmar mais uma vez que o valor do gradiente é 0, quando o $w$ está ali entre próximo de 1! Ou seja, nos confirmando que o $w$ para minimizar nossa função de perda é algo em torno disso. Mas, como já disse, não podemos parar por aí e precisamos de algo mais robusto para chegar à essa conclusão.

Agora, vamos de fato realizar a técnica do gradiente descendente. Eu vou escrever a função comentada, primeiramente em uma versão que eu acredito ser mais simples, e depois tento falar mais ou menos o que foi feito:

```python
def fazer_gradiente_descendente(w_inicial, passo, iteracoes=100):
    
    # Crio uma lista contendo o valor do primeiro argumento:
    w_vals = [w_inicial]
    
    # Inicio um loop pelo número de iterações no terceiro argumento:
    for i in range(iteracoes):
        
        # Aplico a função do gradiente com o último valor de w presente na minha lista:
        grad = gradiente(w_vals[-1], x, y)
        
        # Por último, eu adiciono à lista que eu criei no início o último valor de w porém eu subtraio o valor do gradiente
        # multiplicado ao valor do passo, e é isso que vai permitir que eu "caminhe" pela função!
        w_vals.append(w_vals[-1] - passo*grad)
           
    return w_vals
    
```

```python
w_vals = fazer_gradiente_descendente(0, 0.000001, 5000)

plt.figure(figsize=(10,4))
plt.plot(w_vals)
plt.xlabel('iteracao', fontsize=20)
plt.ylabel('w', fontsize=20)
plt.title('Evolução do Gradiente Descendente\nFinal w: %s'%round(w_vals[-1],5), fontsize=25)
```

Existe uma outra maneira de produzir o gradiente descendente, com um nível de tolerância em vez de um número definido de iterações.

```python
def fazer_gradiente_descendente(w_inicial, passo, tolerancia=0.0001):
    
    w_vals = [w_inicial]
    
    while True:
        
        grad = gradiente(w_vals[-1], x, y)
        w_vals.append(w_vals[-1] - passo*grad)
        
        if (w_vals[-1] - w_vals[-2]) <= tolerancia:
            break
        
            
    return w_vals
    
```
