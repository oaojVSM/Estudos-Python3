# Explorando outros tipos de regressões

Quando estudamos estatística, econometria ou machine learning, um dos primeiros assuntos que aprendemos (pelo menos falando de modelos preditivos), é a regressão linear. E não me entenda mal, a regressão linear padrão pode ser uma ferramenta muito poderosa, e mesmo que não replique idealmente a realidade na maioria das vezes, pode ao menos dar um método para a nossa tomada de decisão na ausência da possibilidade da criação de um modelo mais robusto, por exemplo.

Porém, mesmo apesar de a regressão linear ter o seu valor, ela não deixa de ser um algoritmo bem simples e, querendo ou não, limitado. Justamente por isso, vale muito a pena explorar outros tipos de algoritmos de regressão, como a **regressão linear polinomial**, ou então as **árvores de regressão**, ou até mesmo, a derivação da árvore de regressão que seria a **floresta aleatória de regressão**, e o mais interessante desses algoritmos é que eles não ficam limitados a plotar uma reta, ou seja, a gente começa a abandonar a simplicidade de uma linearidade que é trazida pela regressão linear padrão.

Para isso, pretendo utilizar o mesmo dataset e rodar diferentes modelos de regressão e ver qual que se porta melhor!
