## Iniciando a modelagem:

Primeiramente, para realizarmos a modelagem da regressão linear, teremos que separar as nossas colunas do DF entre as variáveis dependentes e independentes (regressando e regressores, previsto e previsores etc etc...). Sendo que a nossa variável prevista é o ``preco`` e as nossas variáveis previsoras são todas as outras (excluindo o endereço)!

```python
#Portanto, vamos criar dataframes separando essas variáveis:
df.columns
```

```python
# X = previsores
# Y = previsto

X = df[['renda_local', 'idade_imovel_local', 'n_comodos_local', 'n_quartos_local', 'pop_local']]

y = df['preco']
```

Primeiramente, o que queremos fazer é dividir os nossos dados entre a parte que usaremos para testar o modelo e a parte que usaremos para ensinar o modelo, e para isso, utilizaremos uma função do Scikit Learn:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

Bom, vamos falar um pouco sobre o que eu fiz acima:

- Primeiramente, importei a funçao ``train_test_split`` do SKLearn
- Depois, usei tuple unpacking para separar as porções de teste e treinamento dos nossos dados
- Quanto aos argumentos da função, recomendo que veja a documentação, sendo que é relativamente simples!

O que vou fazer agora é importar do SKLearn, o módulo de regressão linear para realizarmos uma regressão com os nossos dados!

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

Agora que vamos realmente realizar nossa regressão por meio da função **.fit()**, na qual fornecemos as variáveis dependentes e independentes!

```python
# Realizando o fit e armazenando na variável model

model = lm.fit(X_train, y_train)
```

Importante: Repare que, para realizar o modelo, estou passando apenas os dados de treinamento, isso porque, quando formos realizar as previsões, iremos utilizar os dados de teste!

```python
# Para ver o intercepto:

print(model.intercept_)
```

```python
# Puxando os coeficientes, temos um array, que está com a respectiva ordem do nosso dataframe:

model.coef_
```

```python
# Aqui coloco os coeficientes em um dataframe, para facilitar a visualização

pd.DataFrame(model.coef_ , X.columns, columns=['coeficientes'])
```

Beleza, conseguimos realizar a regressão, mas **como fazemos previsões com base nisso?**

## Realizando previsões com base no nosso modelo:

```python
# Vamos realizar previsões utilizando as variáveis de reste previsoras!

previsoes = model.predict(X_test)
```

```python
previsoes
```

Esse array que nos foi retornado são os previsões dos preços das casas para cada set de variáveis previsoras!

Lembrando que, temos um outro array que possui os valores corretos das casas, o ``y_test``!

```python
y_test
```

Vamos plotar os dados pra ver se conseguimos ver alguma coisa útil:

```python
sns.scatterplot(y_test, previsoes)
```

Bom, ao olhar nosso plot, parece que fizemos um trabalho decente de previsão, visto que há uma clara relação entre os dados!

Outra forma de visualizar como nosso modelo performou seria por meio de um histograma dos resíduos, para ver a região dos valores nos quais eles se concentram:

```python
sns.histplot((y_test-previsoes), kde=True)
```

Vejamos que interessante! Os nossos resíduos parecem seguir uma distribuição normal (indicação de um bom modelo) em torno de 0, sendo que a maior parte dos nossos valores está perto do 0! Então, isso está novamente nos indicando que fizemos um trabalho decente!

Agora vamos calcular algumas das métricas que expliquei no notebook anterior!

```python
from sklearn import metrics as mtrc
```

```python
mtrc.mean_absolute_error(y_test, previsoes)
```

```python
mtrc.mean_squared_error(y_test, previsoes)
```

```python
np.sqrt(mtrc.mean_squared_error(y_test, previsoes)) # RMSE
```
