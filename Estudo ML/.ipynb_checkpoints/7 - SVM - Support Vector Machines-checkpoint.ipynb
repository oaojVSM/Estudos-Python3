# Support Vector Machines

Bem vindo ao Notebook sobre *Support Vector Machines*, um algoritmo de classificação *(também tem uma aplicação de regressão, mas aqui me limitarei a falar sobre sua aplicação em classificação)*. Neste notebook tentarei explicar um pouco da lógica por trás (*digo um pouco, por que SVM possui muita matemática complicada por trás*) e mostrar sua aplicação com o *scikit-learn*!

## O que é uma Support Vector Machine

SVM é um algoritmo de ML, de aprendizado supervisionado, que analisa os dados de entrada, cria uma separação espacial entre esses pontos de amostra, e classifica novos pontos de acordo com a sua posição relativa a separação criada.

Vamos tentar pensar em um exemplo bem básico, para tentar tornar isso lúdico:

```python
# Importando as libs básicas para exemplificar o que eu quero
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_exemplo = pd.read_excel('Peso_Esquilos.xlsx') # Carregando uma amostra artifical, bem simples, que criei no Excel
```

```python
# Plotando a amostra
sns.scatterplot(data=df_exemplo,
               x=df_exemplo['Peso'],
               hue=df_exemplo['Class'],
               y=0) 
```

Acima eu plotei uma amostra artificial que eu criei (está disponível no github) que contém o peso de 12 esquilos, a classificação deles entre obeso e não obeso e o ponto médio dos pesos dos esquilos da amostra. E a intuição que estou tentando trazer por meio de plotar essa amostra é a seguinte:

Podemos, para classificar um novo ponto, usar o ponto médio essa distribuição como um "separador" para saber se o esquilo é obeso ou não. E, se o peso de um novo esquilo for maior do que o do ponto médio, diremos que ele é obeso, e se não for maior do que o ponto médio, diremos que ele não é obeso!

De forma *extremamente simplificada*, essa é a lógica por trás do SVM!

### Entrando um pouco mais no detalhe...

Na verdade na verdade, o que eu mostrei acima é a lógica do SVC, ou **Support Vector Classifier**, mostrando, na verdade, um **Classificador de Máxima Margem** *(Maximum Margin Classifier)*, pois estamos falando do ponto médio entre os pontos extremos da distribuição.

Ok, eu sei que simplesmente joguei um monte de terminologia pra fora sem realmente explicar do que estou falando, vamos com calma, vou tentar explicar brevemente do que se trata cada um desses termos e, deixarei alguns materiais de referência no final para caso queira entrar no detalhe de cada um desses assuntos.

Este ponto médio que trouxe acima, se trata do **classificador**, e existem alguns tipos de classificadores. No caso, aquele ponto laranja é um **Maximum Margin Classifier**, porque ele é o ponto médio entre os dois pontos extremos mais próximos ao centro da distribuição, ou seja, ele maximiza a distância entre as duas classificações da distribuição.

Porém, o **Maximum Margin Classifier** é **péssimo** para lidar com outliers, já que ele pega os pontos extremos, ou seja, se tivéssemos um esquilo com peso alto e que não fosse obeso (as vezes esse esquilo é comprido, sei lá), o ponto médio entre as duas classificações pode ficar meio esquisito, ou ainda pior, se esse outlier estivesse "no meio" do peso dos ratos obesos, a divisão iria ficar toda zoada.

Para solucionar isso, criamos um outro tipo de separador, um com **Soft Margins** que abre espaço para a permissão de erros de classificação (ou seja, abrimos mão de uma alta variância na nossa classificação, para ter um pouco mais de viés), evitando um overfit. Esse vetor de **soft margins** nada mais é do que o ponto médio entre pontos que não são os extremos da nossa amostra, até achar um separador que seja ótimo, e este ponto ótimo é encontrado através da **validação cruzada**.

Esse **Soft Margins Classifier** também é chamado de **Support Vector Classifier**, e este nome vem do fato que as observações usadas para criar o classificador são chamadas de vetores de suporte!

### SVC

O SVC, quando estamos falando de uma amostra uni-dimensional, se trata de um ponto. Quando estamos falando de amostras bi-dimensionais, esse ponto passa a ser uma linha, para amostra tri-dimensionais, estamos falando de planos, e para 4 dimensões em diante, são hiperplanos.

*Na verdade, todos os classificadores são hiperplanos, mas começamos a chamar de hiperplanos de fato quando falamos a partir de 4 dimensões, porque antes disso temos nomes específicos para os hiperplanos*

### O que de fato é SVM?

Bom, agora que você sabe mais ou menos a lógica por trás de como o algoritmo de **Support Vector Machines** faz as classificações, posso tentar explicar o que de fato é o SVM...

O que de fato faz a classificação é o SVC, o classificador, mas o SVM é a lógica que vai criar esse classificador, e faz isso com umas matemáticas bem complicadas, que não vou detalhar aqui. Mas vou jogar alguns conceitos chave.

Para entender a necessidade para essas matemáticas complicadas, imagine que nem todas as distribuições que usaremos para criar um classificador serão tão *bonitinhas* que nem a do peso dos esquilos, podemos ter distribuições em que uma classificação se concentra no centro, e uma classificação se concentra nos extremos. Se tivermos uma distribuição dessas, não adianta simplesmente tentar colocar um classificador em qualquer lugar que nunca conseguiremos dividir a amostra com eficácia.

E é pra isso que existe o SVM! Ele vai usar essas matemáticas complicadas para mexer com a **dimensionalidade** da nossa amostra, até conseguir encontrar um classificador que vai separar as observações de maneira ótima!

A matemática complicada fica nas metodologias que existem para ir aumentando a dimensionalidade e calcular a relação entre os diferentes pontos da nossa amostra. A metodologia mais comum, que é a padrão do *scikit learn* é a **Radius Basis Function**, ou RBF. Mas, novamente, a intuição para encontrar a dimensionalidade ideal, também reside na validação cruzada.

*Não vou entrar no detalhe da RBF (porque, sinceramente, não sei tanto à respeito) mas deixarei materiais para referência no final do notebook*.

## Colocando em prática

Espero ter conseguido dar uma breve noção do que são SVMs e SVCs, e agora, vamos tentar colocar isso em prática. Também vou mostrar como aplicar algumas outras metodologias interessantes como o `GridSearchCV` e **normalização**.

```python
# Importando os pacotes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler # Objeto que faz a normalização de uma tabela
from sklearn.svm import SVC # O algoritmo de ML
from sklearn.model_selection import GridSearchCV # Algoritmo que permite realizar a validação cruzada para escolher o modelo
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix # Métricas para avaliação do modelo
```

Para aplicar o SVM, utilizarei um dataset padrão do Scikit Learn!

Para demonstrar a aplicação do SVM, utilizarei um dataset de um repositório famoso de dados para projetos de Machine Learning, o [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/index.php), especificamente, utilizarei o dataset [*default of credit card clients*](https://archive.ics.uci.edu/ml/datasets/default+of+credit+card+clients#).

Eu fiz o download do dataset em Excel e vou importá-lo:

```python
df = pd.read_excel('default of credit card clients.xlsx')
df.head()
```

Apenas para efeitos de demonstração, vou mostrar o que aconteceria se simplesmente tentássemos rodar um SVM diretamente nesse dataset sem fazer nenhum tipo de análise exploratória ou ajustes na tabela ou modelo. Apenas pegando esse dataset e colocando o algoritmo padrão do `sklearn` pra rodar:

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('default payment next month', axis=1), 
                                                    df['default payment next month'], 
                                                    test_size=0.33, 
                                                    random_state=101)
```

```python
%%time
modelo_svc = SVC()
modelo_svc.fit(X_train, y_train)
previsao = modelo_svc.predict(X_test)
plot_confusion_matrix(modelo_svc,
                     X_test,
                     y_test)
print(classification_report(y_test,previsao))
```

Veja só o que aconteceu... O nosso modelo conseguiu uma acurácia de 78% porém isso não deve indicar de forma alguma algo positivo, simplesmente porque classificou todos como 0! Nosso modelo não classificou nenhuma observação como 1! Veja no `classification report` que o recall e precisão para class 1 foi 0! E inclusive o `sklearn` nos deu um aviso falando que o F-score foi mostrado como 0, porém na verdade, não pode ser calculado porque não houveram observações previstas com valor 1.

## Fazendo do jeito certo...

Ok, agora que consegui mostrar que você jamais deve fazer o que eu demonstrei acima, e isso não vale só para SVM mas para qualquer tipo de modelo, vou mostrar o jeito certo de fazer as coisas, e vamos começar com uma análise exploratória...

```python
# Nem todo mundo faz isso, mas eu gosto muito de começar a ter uma visão dos dados por meio do describe:
df.describe()
```

Para fazer do jeito certo, vamos olhar coluna por coluna:
- **ID**: Assumo que se trata de uma espécie de código de cliente, e essa informação, por não agregar nada substancial ao nosso modelo, pode acabar nos atrapalhando, então vou dropá-la logo em seguida

- **LIMIT_BAL**: Se trata de uma variável contínua que representa o crédito do cliente.

- **SEX**: Se trata do gênero do cliente (1=homem; 2=mulher), então é uma **variável categórica** que precisaremos transformar em dummy.

- **EDUCATION**: Nos traz o nível de educação do cliente (1 = graduate school; 2 = university; 3 = high school; 4 = others), *olhando o máximo, vemos que tem o número 6 (e no mínimo tem 0), que não é explicado pelo site do UCI, significando que, ou temos um problema nos dados, ou na descrição dos dados, por ora, vamos ignorar isso...*, de toda forma, se trata de info. categórica que precisará ser transformada em *dummy*.

- **MARRIAGE**: Status de casamento do cliente (1 = married; 2 = single; 3 = others), *novamente vemos o mínimo 0*...

- **AGE**: é a idade, e vendo máximo e mínimo, parece que estamos ok quanto a essa coluna...

- **PAY_1...6**: é o histórico de pagamento de 6 meses, e, para cada mês, temos as seguintes classificações: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above.

- **BILL_AMT1...6**: O valor das contas referentes às colunas anteriores.

- **PAT_AMT1...6**: O valor pago das respectivas contas.

- **default payment next month**: É a nossa variável explicativa, se o cliente entrou em *default* no mês seguinte (assumo que 1=default e 0=pagou conta). Nessa coluna temos uma observação interessante, veja que a média é 0,22, ou seja, temos muito mais clientes que pagaram a conta (status 0) do que clientes que não pagaram, mais especificamente 22% não pagaram a conta, e isso explica porque, nosso modelo, mesmo chutando 0 para todo mundo, consegue acertar cerca de 78%!

### Fazendo a limpeza dos dados:

```python
# Primeira coisa a se fazer é dropar a coluna ID, que não agrega em nada e só atrapalha
df.drop(columns='ID', inplace=True)
```

Agora, para os valores que aparecem como 0 nas colunas *MARRIAGE* e *EDUCATION*, como na descrição dos dados, não há nenhuma explicação para o valor 0 aparecer nessas colunas, poderíamos interpretar que seriam dados faltantes em nosso dataset, o que não é ideal.

Vamos dar uma olhada se isso acontece em muitos casos:

```python
len(df.loc[(df['EDUCATION'] == 0) | (df['MARRIAGE'] == 0)])
```

Felizmente, parece que temos apenas 68 linhas onde temos os valores 0 para aquelas colunas, em um dataframe de 30 mil linhas! Então, por via das dúvidas, posso felizmente dropar esses acontecimentos sem prejuízo...

```python
df_ajustado = df.loc[(df['EDUCATION'] != 0) & (df['MARRIAGE'] != 0)]
df_ajustado.describe()
```

### Fazendo o One-Hot Encoding

Como eu disse acima, na nossa breve análise exploratória, temos algumas variáveis categóricas em nosso dataframe, que precisam ser transformadas em *dummy*, e algumas pessoas chamam esse processo de *One-Hot Encoding*. E fazemos isso da seguinte forma:

```python
df_ajustado.columns
```

```python
df_ajustado = pd.get_dummies(df_ajustado,
                            columns=['SEX',
                                    'EDUCATION',
                                    'PAY_0', 
                                    'PAY_2', 
                                    'PAY_3',
                                    'PAY_4',
                                    'PAY_5', 
                                    'PAY_6'],
                            drop_first=True) # Uso o dropfirst para evitar eventuais multicolinearidades, é uma boa prática
```

```python
df_ajustado.describe() # Veja que agora temos muito mais colunas!
```

```python
df_ajustado.head()
```

### Normalização

Agora vamos normalizar os nossos dados. No caso do SVM utilizando a RBF, isso é necessário pois a RBF assume que nossos dados estão seguindo algo parecido com uma distribuição normal com centro em 0 e desv. pad. de 1. E isso pode ser feito por meio do Standard Scaler!

```python
df_ajustado.columns
```

```python
padronizador = StandardScaler()
colunas_a_normalizar = ['LIMIT_BAL', 'MARRIAGE', 'AGE', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3',
       'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2',
       'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'] # Aqui trago as colunas que não são binárias
padronizador.fit(df_ajustado[colunas_a_normalizar])
```

```python
padronizador.scale_
```

```python
padronizador.mean_
```

```python
padrao = padronizador.transform(df_ajustado[colunas_a_normalizar])
df_padronizado = pd.DataFrame(padrao, columns=df_ajustado[colunas_a_normalizar].columns)
df_padronizado.head()
```

```python
#Trazendo de volta as colunas categóricas:
df_padronizado = pd.concat([df_padronizado.reset_index(), df_ajustado.drop(columns=colunas_a_normalizar).reset_index()], axis=1)
```

```python
df_padronizado.describe()
```

### Downsampling

Agora o nosso dataset está quase pronto para rodar o modelo! O próximo passo é fazer um downsampling, e vou fazer isso por dois motivos:
- Primeiro: Vai permitir que realizemos o balanceamento dos dados, ou seja, que tenhamos a mesma quantidade de 1s e 0s da nossa variável resposta.
- Segundo: Nosso dataset é relativamente grande (30k linhas) o suficeiente para que não precisemos usar ele inteiro para treinar o modelo, então, diminuir o seu tamanho vai permitir rodar a validação cruzada (um dos próximos passos) muito mais rapidamente!

O primeiro passo vai ser dividir o nosso dataframe em 2, um só com 0s para a variavel resposta, e um só com 1s:

```python
coluna_resposta = 'default payment next month'
df_default = df_padronizado.loc[df_padronizado[coluna_resposta] == 1].drop(columns='index')
df_sem_default = df_padronizado.loc[df_padronizado[coluna_resposta] == 0].drop(columns='index')
```

```python
df_default[coluna_resposta].unique()
```

```python
df_sem_default[coluna_resposta].unique()
```

Agora nosso DF está separado em 2, um só com os clientes que deixaram de pagar, e um só com os clientes que pagaram em dia. O próximo passo é fazer o *downsample* de cada um desses DFs e juntá-los

```python
df_default_downsampled = df_default.sample(n=1000, # Vou pegar 1000 observações
                                           replace=False,
                                           random_state=42
                                          ).reset_index().drop(columns=['index'])
df_default_downsampled.head()
```

```python
df_sem_default_downsampled = df_sem_default.sample(n=1000,
                                                  replace=False,
                                                  random_state=42
                                                  ).reset_index().drop(columns=['index'])
df_sem_default_downsampled.head()
```

```python
df_downsampled = df_default_downsampled.append(df_sem_default_downsampled)
df_downsampled.describe()
```

### Fazendo um novo train-test-split

Agora que o dataset está pronto, com todos os ajustes necessários, podemos fazer um novo `train-test-split`

```python
X_train, X_test, y_train, y_test = train_test_split(df_downsampled.drop('default payment next month', axis=1), 
                                                    df_downsampled['default payment next month'], 
                                                    test_size=0.33, 
                                                    random_state=42)
```

### Teste novo do modelo, que ainda não está ideal!

Podemos agora, testar de novo o modelo:

```python
%%time
modelo_svc = SVC()
modelo_svc.fit(X_train, y_train)
previsao = modelo_svc.predict(X_test)
plot_confusion_matrix(modelo_svc,
                     X_test,
                     y_test)
print(classification_report(y_test,previsao))
```

Veja que agora nosso modelo começou a nos dar um resultado extremamente melhor! Vamos comparar com como ele se portou no treino:

```python
previsao_treino = modelo_svc.predict(X_train)
plot_confusion_matrix(modelo_svc,
                     X_train,
                     y_train)
print(classification_report(y_train,previsao_treino))
```

Veja também que a performance foi bem parecida no treino e no teste, mostrando uma ausência de *overfitting*!

Porém, você deve ter visto que eu disse que nosso modelo ainda não está ideal. Isso porque nos utilizaremos da **validação cruzada** para tentar melhorá-lo!

### Validação Cruzada para montar o modelo ideal

Agora nos utilizaremos da validação cruzada, por meio de um algoritmo do `scikit-learn` chamado `GridSearchCV`!

Primeiro vou definir quais são os parâmetros a serem testados:

```python
parametros = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001, 'scale']} 
```

```python
%%time
cross_val = GridSearchCV(SVC(), parametros, refit=True, scoring='accuracy')
cross_val.fit(X_train,y_train)
```

```python
print(cross_val.best_params_)
```

Veja que interessante, por meio da validação cruzada, foi possível concluir que os melhores parâmetros do SVM para o nosso dataset são de `{'C': 10, 'gamma': 0.001}`! E vamos rodar novamente o modelo usando esses parâmetros e ver se conseguimos alguma melhora!

### Criando um novo modelo otimizado:

Agora que rodamos a validação cruzada, podemos verificar se conseguimos melhorar o nosso modelo em relação ao anterior, vamos treinar um modelo novo com os parâmetros retornados:

```python
%%time
modelo_svc_CV = SVC(C=10, gamma=0.001)
modelo_svc_CV.fit(X_train, y_train)
previsao = modelo_svc_CV.predict(X_test)
plot_confusion_matrix(modelo_svc_CV,
                     X_test,
                     y_test)
print(classification_report(y_test,previsao))
```

O que podemos fazer agora, seria utilizar ambos os modelos em nossa base original completa e ver qual se porta melhor!

```python
# Utilizando o modelo com a Validação Cruzada:

previsao = modelo_svc_CV.predict(df_padronizado.drop(columns=[coluna_resposta, 'index']))
plot_confusion_matrix(modelo_svc_CV,
                     df_padronizado.drop(columns=[coluna_resposta, 'index']),
                     df_padronizado[coluna_resposta])
print(classification_report(df_padronizado[coluna_resposta],previsao))
```

```python
# Modelo sem a validação cruzada:

previsao = modelo_svc.predict(df_padronizado.drop(columns=[coluna_resposta, 'index']))
plot_confusion_matrix(modelo_svc,
                     df_padronizado.drop(columns=[coluna_resposta, 'index']),
                     df_padronizado[coluna_resposta])
print(classification_report(df_padronizado[coluna_resposta],previsao))
```
