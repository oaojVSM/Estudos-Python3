## Distribution Plots

Primeiramente, vamos falar de distribution plots:

```python
import seaborn as sns
```

Vamos trabalhar aqui com uma base de dados genérica que já vem junto com o Seaborn e podemos importar como DataFrame, no caso, a base que utilizaremos chama "Tips". <br><br>

Para importá-la, basta utilizar:

```python
tips = sns.load_dataset('tips')
```

```python
tips.info() #veja que o tips é um dataframe!
```

```python
import pandas as pd #portanto importaremos o pandas e o numpy
import numpy as np
```

```python
tips
```

Para facilitar, irei renomear as colunas do DataFrame para português!

```python
gorjetas = tips
```

```python
gorjetas.rename(columns={
    'total_bill' : 'valor_total',
    'tip' : 'gorjeta',
    'sex' : 'sexo',
    'smoker' : 'fumante',
    'day' : 'dia',
    'time' : 'hora',
    'size' : 'qtd_pessoas'
}, inplace=True)
```

```python
gorjetas
```

Pronto, agora que temos nossa base de dados, vmaos começar a visualizá-la utilizando o Seaborn!

Primeiramente, vamos falar do histplot, que nada mais é do que um tipo de histograma, veja e entraremos em maiores detalhes mais pra frente:

### Histplot (histograma simples)

```python
sns.histplot(tips.valor_total, kde=True)
```

Aqui podemos ver que temos um histograma que possui uma KDE (Kernel Density Estimation, ou Estimativa de Densidade Kernel, em portguês), que é uma forma de você estimar uma distribuição contínua para uma base de dados finita, ou seja, em vez de uma histograma, teríamos uma distribuição suave e contínua! [Veja mais aqui!](https://pt.wikipedia.org/wiki/Estimativa_de_densidade_kernel)

Se quisermos, podemos plotar apenas o histograma, e também podemos mudar o número de bins (a quantidade de barras que aparece no histograma)!

```python
sns.histplot(tips.valor_total, kde=False, bins=50)
```

### jointplot

Uma outra plotagem interessante que temos dentro do Seaborn é o **jointplot**, que é uma boa maneira de verificar dados com duas variáveis. Esse plot é bem interessante quando tratamos de regressões lineares:

```python
sns.jointplot(data=gorjetas,x='valor_total', y='gorjeta')
```

O jointplot suporta um parâmetro chamado "kind", que nada mais é do que o tipo de plot. Veja que é um parâmetro que não especifiquei no plot acima, então, por padrão ele é atribuído para o tipo 'scatter', veja:

```python
# Mesmo comando acima porém com o argumento Scatter

sns.jointplot(data=gorjetas,x='valor_total', y='gorjeta', kind='scatter')
```

Podemos também utilizar outros tipos, veja um que pode ser considerado bonito mas não tão funcional assim:

```python
sns.jointplot(data=gorjetas, x='valor_total', y='gorjeta', kind="hex")
```

Podemos plotar também um tipo com uma linha de regressão:

```python
sns.jointplot(data=gorjetas, x='valor_total', y='gorjeta', kind="reg")
```

### pairplot

Um outro tipo de plotagem que é interessante para distribuições é o **pairplot**, que toma o DF como argumento e automaticamente estabelece relações entre todas as colunas numéricas. Aqui também falaremos do argumento **hue**, que é um parâmetro que utiliza as colunas categóricas!

```python
# Essa função pode demorar um pouco para rodar quando temos muitos dados
sns.pairplot(gorjetas)
```

Veja que essa função nos retornou uma espécie de matriz, relacionando de forma bidimensional as diferentes variáveis do nosso DF, e quando as relações são das variáveis com elas mesmas, possuímos um histograma (diagonal principal da matriz)!

Veja como funciona o argumento **hue**, que é um argumento que pode ser inserido em diferentes plotagens, sendo que irá classificar os dados de acordo com alguma coluna categórica do nosso DF, no caso, vamos usar o dado  "fumante":

```python
sns.pairplot(gorjetas,hue='fumante')
```

```python
sns.pairplot(gorjetas,hue='sexo')
```

## Plots Categóricos

Agora um pouco de plots categóricos, que vão nos mostrar informações baseados nas colunas de categoria que tivermos em nosso DF

O mais simples de todos é o countplot, que nos mostra a quantidade de acordo com a categoria:

```python
sns.countplot(data=gorjetas, x='fumante')
```

Temos também os **barplots**, que mostram alguma estatística (que podemos escolher), para as categorias do nosso DF:

```python
#Por default ele nos retorna a média:
sns.barplot(data=gorjetas, x='sexo', y='valor_total')
```

```python
#Mas podemos escolher a estatística
sns.barplot(data=gorjetas, x='sexo', y='valor_total', estimator=np.std)
#No caso, estou plotando o Desv. Pad.
```

### Boxplot e Violinplot

Algumas plotagens um pouco mais complexas mas muito interessantes são os **boxplots** e os **violinplots**, para saber mais sobre a interpretação de boxplots, [clique aqui!](https://www.simplypsychology.org/boxplots.html#:~:text=When%20the%20median%20is%20in,positively%20skewed%20(skewed%20right).)

```python
sns.boxplot(data=gorjetas, x='dia', y='valor_total')
```

```python
#Podemos adicionar o argumento hue também
sns.boxplot(data=gorjetas, x='dia', y='valor_total', hue='sexo')
```

Uma plotagem que é tem uma função bem semelhante ao boxplot é o **violinplot** ou, gráfico em violino, que nos mostra a função de distribuição do dado, de forma espelhada. Acredito que isso fique mais simples de entender visualizando:

```python
sns.violinplot(data=gorjetas, x='dia', y='valor_total')
```

```python
#Novamente podemos colocar o hue
sns.violinplot(data=gorjetas, x='dia', y='valor_total', hue='sexo')
```

E podemos colocar um argumento interessante quando colocamos o hue, que é o **split**, veja como funciona:

```python
sns.violinplot(data=gorjetas, x='dia', y='valor_total', hue='sexo', split=True)
```

### Stripplot

Um **stripplot** é um irmão do scatterplot, porém ele é utilizado quando uma das variáveis nos eixos é categórica:

```python
sns.stripplot(data=gorjetas, x='dia', y='valor_total', hue='sexo')
```

Uma plotagem parecida com o strip é o **swarmplot**, que, no caso, se diferencia do strip pelo fato de no swarm, os pontos não poderem ficar um em cima do outro, ou seja, não há overlapping, mas adianto que essa plotagem **não é boa quando é uma quantidade grande de dados**, tanto pelo fato da visualização ficar extremamente poluída, quanto pelo fato de ser necessário um enorme poder computacional para gerar este gráfico com grandes bases de dados, mas, como nossa base é pequena, pode ser utilizado:

```python
sns.swarmplot(data=gorjetas, x='dia', y='valor_total', hue='sexo')
```

## Plots Matriciais

Podemos falar também em plots matriciais, que nada mais são do que **heatmaps**. Para isso, vou importar uma outro DF do Seaborn, que mostra dados de voos aéreos nos EUA.

```python
voos = sns.load_dataset('flights')
```

```python
voos
```

```python
voos.rename(columns={
    'year' : 'ano',
    'month' : 'mes',
    'passengers' : 'qtd_passageiros'
}, inplace=True)
```

```python
voos
```

Vamos tentar realizar uma plotagem de heatmap com os dados dessa forma para ver o que acontece:

```python
sns.heatmap(voos)
```

Veja que nos foi retornado um erro! Isso acontece porque, para conseguirmos plotar heatmaps, precisamos que os nossos dados estejam em uma forma matricial!!! Isso pode ser feito por meio de uma matriz de correlação, ou então por meio de uma Pivot Table (que é o que faremos a seguir):

```python
voos_pt = voos.pivot_table(values='qtd_passageiros', 
                           index='mes', 
                           columns='ano')
```

```python
voos_pt
```

Agora que nossos dados estão em uma forma matricial, vamos tentar plotar nosso heatmap:

```python
sns.heatmap(voos_pt,cmap='Purples')
```

```python
#Podemos colocar alguns argumentos, como linhas e mudar as cores de nosso plot:
```

```python
sns.heatmap(voos_pt,cmap='coolwarm',linecolor='white',linewidths=1)
```

```python
sns.heatmap(voos_pt,cmap='flare',linecolor='white',linewidths=3)
```

```python
sns.heatmap(voos_pt,cmap='mako',linecolor='black',linewidths=1)
```

## Grades

Com grades, ou grids, podemos criar tipos de pairplots, que vimos anteriormente, que podem ser customizados da forma que quisermos, veja:

```python
#Vamos utilizar uma outra base de dados:
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
#Usando o pairplot normal:
sns.pairplot(iris)
```

```python
#adicionando outros argumentos
sns.pairplot(iris,hue='species',palette='rainbow')
```

Agora, vamos customizar esse pairplot, utilizando o **PairGrid**:

```python
# Just the Grid
sns.PairGrid(iris)
```

```python
# Aí podemos colocar algum tipo de plot que quisermos no grid
g = sns.PairGrid(iris)
g.map(sns.scatterplot)
```

Podemos colocar diferentes plots nos diferentes lugares da grid, como se fosse uma matriz!

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

## Plots de Regressão

Aqui falaremos de como plotar gráficos de regressão:

```python
# Vamos voltar a utilizar nosso DF de gorjetas:
sns.lmplot(data=gorjetas, x='valor_total', y='gorjeta')
```

```python
# Podemos colocar hue também
sns.lmplot(data=gorjetas, x='valor_total', y='gorjeta', hue='sexo')
```

```python
# Podemos customizar bastante o gráfico, e o mecanismo é igual ao matplotlib
sns.lmplot(x='valor_total',y='gorjeta',data=gorjetas,hue='sexo',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':20})
```

Podemos separar os gráficos com base no hue, colocando um gráfico do lado do outro, colocando o argumento que colocaríamos no hue, em **col**, de colunas:

```python
sns.lmplot(x='valor_total',y='gorjeta',data=gorjetas,col='sexo')
```

```python
#Podemos colocar linhas também:
sns.lmplot(x='valor_total',y='gorjeta',data=gorjetas,col='sexo',row='dia')
```

```python
#E podemos voltar a colocar o hue, se quisermos!
sns.lmplot(x='valor_total',y='gorjeta',data=gorjetas,col='sexo',row='dia',hue='hora')
```

## set_style

Uma função interessante do Seaborn é a **set_style**, que podemos utilizar para mudar o estilo de nossos gráficos, veja:

```python
sns.set_style('darkgrid')
#Vou plotar o mesmo gráfico de cima para vermos a diferença:
sns.lmplot(x='valor_total',y='gorjeta',data=gorjetas,col='sexo',row='dia',hue='hora')
```

Importante mencionar que isso valerá para todas as plotagens seguintes!
