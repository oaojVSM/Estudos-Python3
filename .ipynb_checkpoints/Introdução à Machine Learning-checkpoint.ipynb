- Mas, primeiro, o que é **Machine Learning**?

Machine Learning é um método de análise de dados que automatiza a construção do modelo analítico. Ou seja, por meio de algoritmos que aprendem por meio dos dados, a metodologia de ML (Machine Learning) permite que o computador encontre insights e saiba realizar a modelagem sem que o modelo seja programado explicitamente!<br><br>

Outro assunto que é encontrado muito quando se falando de ML, são as redes neurais.

- Mas, o que são **redes neurais**?

Redes neurais são uma forma de modelar sistemas de "neurônios" matematicamente, ou seja, são um tipo de algoritmo de ML. E entraremos mais em detalhe sobre o funcionamento disso mais a frente.


## Supervised Learning

Existem duas maneiras de se treinar um algoritmo de ML. <br><br>

E o primeiro assunto abordado será **supervised learning** ou melhor, treinamento supervisionado. <br><br>

Algoritmos de treinamento supervisionado são treinamentos utilizando exemplos nos quais já sabemos o output desejado, seria o caso, por exemplo, de treinarmos um algoritmo com dados históricos já classificados. <br><br>

O treinamento supervisionado é comumente utilizado em aplicações nas quais estamos objetivando o uso de dados históricos para predizer eventos futuros.<br><br>

O processo de ML, mais especificamente o que envolve o treinamento supervisionado funciona da seguinte forma:

- Primeiro extraímos os dados e o limpamos (realizamos o ETL)

- Depois separamos os dados em 2 (ou 3) partes:
    - Uma parte (a maior parte) será utilizada para de fato treinarmos o algoritmo
    - Uma outra parte usaremos para validar o algoritmo (essa que pode ser descartada, caso queira pular uma etapa) - isso não é muito recomendável, mas se estiver realizando um algoritmo relativamente simples e não muito importante, pode pular esta etapa
    - E a outra parte dos dados será usada para de fato testar o seu algoritmo (será o primeiro contato do seu algoritmo com o mundo externo)
    
- A etapa acima é a mais demorada e é a principal, sendo que provavelmente será necessário ficar revisitando o seu algortimo para refiná-lo.

## Avaliação do Modelo

Existem algumas métricas chaves para a avaliação de um modelo, e estas são:

- Acurácia
- Recall (Revocação)
- Precisão
- F1-Score (Pontuação F1)

Mas, primeiramente, já devo observar que, não existem pontuações universalmente boas para cada um dessas métricas e isso depende muito do modelo que está sendo desenvolvido. O que quero dizer é: não tem como falar que uma precisão X pode ser considerada boa, porque dependendo do seu modelo, isso pode variar muito, e explicarei o porquê disso em breve. <br><br>

Para explicar como funcionam essas métricas, tomarei como fato estarmos tratando de um modelo de classificação binária, ou seja, o algoritmo tem que escolher entre 0 e 1. <br> <br>

Para tornar mais intuitivo, vamos supor que nosso algoritmo seja um médico que tenha que decidir se uma pessoa tem ou não uma doença, e é só isso que ele tem que decidir. <br><br>

### Acurácia

A primeira métrica e provavelmente a mais intuitiva que existe quando falamos de avaliar modelagens é a acurácia, e ela é definida simplesmente pela **razão entre a quantidade de previsões corretas que o nosso modelo realizou e a quantidade total de previsões.**<br><br>

Ou seja, se o nosso modelo realizou 80 previsões corretas e realizou um total de 100 previsões, possui uma acurácia de 0,8 ou 80%!<br><br>

Isso pode parecer extremamente intuitivo, **mas vale importantíssima ressalva**, pois a **acurácia só se torna uma boa medida quando a nossa amostra de classificação está bem equilibrada**, ou seja, quando a quantidade de 0s e 1s está em torno de 50% para cada um. Ou seja, quando estamos falando de um modelo que avalia se a pessoa possui uma doença ou não, é provável que a acurácia por si só não seja uma boa métrica, e isso ocorre pelo seguinte motivo:

- Quando estamos falando de verificação da presença de uma doença, é realista assumir que a maior parte das pessoas que passe pela verificação não possua tal doença, ou seja: Se estivermos tratando de uma doença rara, vamos supor que apenas 1% das pessoas tenha apenas, e o nosso modelo está avaliando 100 pessoas.
- **Mesmo se tivermos um modelo inútil**, que avalie todos da mesma forma, simplesmente falando que a pessoa não tem a doença, **ele possuirá uma acurácia de 99%!**

Claro que o exemplo apresentado aqui é exagerado, porém deixa bem claro que **a acurácia por si só pode não ser uma boa métrica, e deve ser considerada com mais peso apenas quando temos amostras bem balanceadas.**

### Precisão

Uma outra métrica importantíssima para avaliar um modelo é a sua precisão, e ela é dada pela razão entre o número de Verdadeiros Positivos, e a soma entre o número de Verdadeiros Positivos e o número de Falsos Positivos: <br><br>


<br><br>

Ou seja, pensando de forma mais intuitiva, é a nossa porcentagem de acerto de positivos, dentro de tudo que classificamos como positivo!


### Recall

Falando de uma outra métrica, temos o Recall (Rechamada ou Revocação), que é dada pela razão entre o número de Verdadeiros Positivos e a soma entre o número de Verdadeiros Positivos e o número de Falsos Negativos:<br><br>



<br><br>

A diferença entre a precisão e o Recall pode ser sutil, mas novamente, tentando pensar de forma mais intuitiva, aqui temos a nossa porcentagem de acerto de positivos dentro de tudo que é positivo na nossa amostra, e não dentro do que classificamos como positivo! <br><br>

Uma métrica pune a classificação de Falsos Positivos e a outra pune a classificação de Falsos Negativos!


### Trade Off entre Recall e Precisão

Geralmente **existe um *trade off* entre recall ou precisão**. Ou seja, precisamos pensar se o nosso modelo deve focar em consertar o Erro do tipo 1 ou tipo 2. E pensemos no exemplo da detecção de doenças novamente. <br><br>

Quando estamos realizando a detecção de uma doença, **faz muito mais sentido realizarmos mais erros do tipo 1 do que do tipo 2!** Pelo seguinte motivo: é melhor você falar que o paciente tem a doença, e realizar maiores diagnósticos quando ele não tem a doença, do que você falar que o paciente não tem a doença (e ele não ir atrás) quando ele realmente tiver! <br><br>

Então, neste caso, preferimos ter Falsos Positivos! Portanto o nosso modelo acaba tendo uma **menor precisão, porém um maior recall!** Justamente porque, dentro de tudo que classificamos como positivo, provavelmente teremos menos casos positivos de fato (diminuindo a precisão), MAS, dentro de tudo que for positivo na amostra, estaremos captando uma maior quantidade de amostras de fato positivas (maior recall)!


*Fonte: [Wikipedia](https://pt.wikipedia.org/wiki/Precis%C3%A3o_e_revoca%C3%A7%C3%A3o)*

### F1-Score

Tendo entendido o que é Recall e Precisão, temos uma métrica que combina esses dois, o **F1-Score**, dado pela média harmônica entre as duas métricas mencionadas:



O objetivo desta métrica é tentar encontrar o equilíbrio ideal entre Precisão e Recall e sua utilização ficará mais clara quando entrarmos em partes mais práticas de ML!
