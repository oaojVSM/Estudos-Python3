# Regressão Logística

Agora falaremos da regressão logística, que é um tipo de regressão que nos ajudará com classificações!

Quando falamos em regressão linear, estávamos prevendo valores contínuos. Agora com a regressão logística, estaremos prevendo classificações! Ou seja, com base em valores que estabeleceremos como regressores, conseguiremos classificar algo (ex. de tentar diagnosticar a presença de uma doença).

Mas o que é a regressão logística? <br><br>

Bom, esse nome vem da [função logística](https://pt.wikipedia.org/wiki/Fun%C3%A7%C3%A3o_log%C3%ADstica) (também chamada de função sigmoide), se tratando de uma função que **sempre**, independente dos valores inputados, variará entre 0 e 1:

Justamente por causa dessa vairação entre 0 e 1, que podemos usar uma função como essa para classficar alguma coisa! Podemos estabelecer algum parâmetro como, se o output da função for maior que 0,5, classificamos como 1, ou se for menor do que 0,5, classificamos como 0 (pensando em uma classificação binária)!

Caso se interesse pelos motivos matemáticos de tudo isso, sinta-se livre para pesquisar mais sobre a função sigmóide!

Para tudo isso ficar mais lúdico, irei passar para apliações da regressão logística em Python!

Bom, agora que fizemos uma análise exploratória dos dados, podemos seguir...

O que faremos agora é tentar lidar com a coluna de idade, que possui dados faltantes!

```python
sns.boxplot(x='Pclass', y="Age", data=treino)
```

Por meio desse gráfico, podemos ver como a idade dos passageiros são distribuídas através das classes. E o que podemos fazer para preencher os valores nulos, é pegar a média da idade dos passageiros por classe e preencher os valores nulos da idade com a média respectiva à classe do passageiro!

*Algo que poderia ser feito que seria até melhor, seria criar um modelo preditivo da idade baseado em todos os fatores...*

```python
def imputar_idade(cols):
    Age = cols[0]
    Pclass = cols[1]
      
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            #return treino[(Pclass == 1)].Age.mean()
            return 37
        elif Pclass == 2:
            #return treino[(Pclass == 2)].Age.mean()
            return 29
        else:
            #return treino[(Pclass == 3)].Age.mean()
            return 24
    else: 
        return Age
```

```python
treino.Age = treino[['Age', 'Pclass']].apply(imputar_idade, axis=1)
```

```python
sns.heatmap(treino.isnull())
```

Veja que agora não temos mais dados em branco na coluna de idades!

Algo que você pode começar a se perguntar é por que agora não partimos para preencher a coluna `Cabin` e o motivo simplesmente é que temos muitos dados em branco para que possamos fazer um preenchimento de forma coerente! Portanto, optarei por deixar esta coluna de lado nas análises.

Inclusive, irei dropar a coluna

```python
treino.drop(
    'Cabin',
    axis=1,
    inplace=True)
```

```python
treino
#Veja que não temos mais a coluna de cabine
```

E, na verdade, irei dropar todas as linhas remanescentes que possuírem algum valor em branco...

```python
treino.dropna(axis=0, inplace=True)
```

```python
sns.heatmap(treino.isnull())
```

Agora, temos que fazer outro tratamento em nosso DF para que possamos de fato começar a lidar com variáveis classificatórias!

## Criando variáveis Dummy

Vamos dar uma olhada no nosso DF, especificamente na coluna de sexo:

```python
treino[['PassengerId', 'Sex']]
```

Veja que o gênero aqui está sendo classificado em `strings` e o nosso algoritmo não saberá interpretar isso, então o que precisamos fazer é criar uma chamada **variável dummy** para que transformemos essa String em algo que nosso algoritmo consiga interpretar! E precisaremos fazer isso em todas as colunas que forem variáveis classificatórias e estiverem escrito como String!

Podemos fazer isso para a coluna `Sex` da seguinte forma:

```python
# Veja a função do Pandas chamada get_dummies:

pd.get_dummies(treino.Sex)
```

Veja o que temos acima! Foi nos retornado um DF que está dizendo, por linha, o gênero da pessoa, sendo que 1 é um caso positivo e 0 é negativo para o gênero... <br><br>

E, neste caso, uma coluna é um preditor perfeito da outra, ou seja, se a pessoa é 0 para female, necessairamente será 1 para male e vice-versa...

Portanto, precisamos só de uma dessas colunas, parte porque só uma irá satisfazer plenamente a nossa necessidade de saber o gênero. E segundo que, se introduzirmos as duas colunas em nosso modelo, causaremos um problema de [multicolinariedade](https://pt.wikipedia.org/wiki/Multicolinearidade#:~:text=Multicolinearidade%20consiste%20em%20um%20problema,lineares%20exatas%20ou%20aproximadamente%20exatas.)!

Enfim, para de fato colocar a variável dummy de sexo, farei o seguinte:

```python
sexo_dummy = pd.get_dummies(treino.Sex, drop_first=True)
```

```python
sexo_dummy
```

```python
embarked_dummy = pd.get_dummies(treino.Embarked, drop_first = True)
```

```python
embarked_dummy
```

Criei dois dataframes, com variáveis dummy para as que a gente vai precisar trocar, e abaixo vou concatenar todos esses DFs:

```python
treino = pd.concat([treino, sexo_dummy, embarked_dummy], axis=1)
```

```python
treino
```

Agora que eu tenho as variáveis dummy, posso dropar as outras que estão classificando como String, ou outras colunas que não vou usar:

```python
treino.drop(['Sex', 'Embarked', 'Name', 'Ticket', 'PassengerId'], axis=1, inplace=True)
```

```python
treino
```

Agora limpamos o nosso DF totalmente e temos um exemplo bom de tabela para realizar uma modelagem!

## Começando a modelagem:

Primeiro vamos definir as variáveis dependentes e independentes:

```python
X = treino.drop('Survived', axis = 1)
y = treino.Survived # o que queremos prever
```

```python
from sklearn.model_selection import train_test_split
```

Separar os dados entre treino e teste:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101 )
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(max_iter=10000)
```

Vamos "fitar" o modelo em nossos dados:

```python
logmodel.fit(X_train, y_train)
```

Aqui vamos testar o modelo:

```python
previsoes = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

Podemos utilizar essa função do SkLearn que nos traz métricas de acurácia, recall e precisão, assim como o F-1 Score da nossa regressão:

```python
print(classification_report(y_test, previsoes))
```

```python
from sklearn.metrics import confusion_matrix
```

Podemos também printar uma matriz de confusão:

```python
confusion_matrix(y_test, previsoes)
```

E esses aqui são os valores previstos (output binário)!
